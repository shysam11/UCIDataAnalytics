```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Display first 5 rows of purchase_data
purchase_data.head()
```

```python
#Count of each column data
purchase_data.count()
```

```python
#Display SN column values
purchase_data["SN"].head()
```

```python
#Players Count - Display the total number of players
Tot_Players=purchase_data["SN"].count()
Tot_Players 
```

```python
#Players Count - Display the total number of players in 2D
Totplyrs_table = pd.DataFrame({"Total Players": [Tot_Players]})
Totplyrs_table
```

```python
#Check the datatypes of the columns
purchase_data.dtypes
```

```python
# Purchase Analysis (Total) - Analysis
#Calculate the number of unique items 
itm_count = len(purchase_data["Item Name"].unique())

# Calculate the average price with format applied
#purchase_data["Price"] = purchase_data["Price"].map("${:.2f}".format)
avg_price = purchase_data["Price"].mean()
#file_df["avg_cost"] = file_df["avg_cost"].map("${:.2f}".format)

# Calculate the number of purchases 
no_of_pur = purchase_data["Purchase ID"].count()

# Calculate the total revenue with format applied
tot_rev = purchase_data["Price"].sum()
```

```python
## Purchase Analysis (Total)
# Display summary table for the above into a DataFrame
purchanly_tot_table = pd.DataFrame({"Number of Unique items": itm_count,
                              "Average price": [avg_price],
                              "Number of Purchases": [no_of_pur],
                              "Total Revenue": [tot_rev]})
purchanly_tot_table
```

```python
#Gender Demographics - Analysis
totg_counts = purchase_data["Gender"].value_counts()

#gen_demo_table = pd.DataFrame({"Percentage of Players": [(totg_counts/Tot_Players)*100],
#                              "Total Count": [totg_counts]})

gen_demo_df = pd.DataFrame({"Total Count": totg_counts,
    "Percentage of Players": (totg_counts/Tot_Players)*100})
#Reorganize the gender demographics
gendemo_table = gen_demo_df[["Percentage of Players","Total Count"]]

```

```python
#Gender Demographics
gendemo_table.head()
```

```python
#Purchasing Analysis(Gender) - Analysis
gender_group = purchase_data.groupby("Gender")
purcnt_grpg = gender_group["Item Name"].count()
avgpur_grpg = gender_group["Price"].mean()
totpur_grpg = gender_group["Price"].sum()

purch_analy_gendemo_table = pd.DataFrame({"Purchase Count": purcnt_grpg,
                                            "Average Purchase Price": avgpur_grpg.map("${:.2f}".format),
                                            "Total Purchase Value": totpur_grpg.map("${:.2f}".format),
                                         "Avg Purchase Total per Person": avgpur_grpg.map("${:.2f}".format)})

#Purchasing Analysis(Gender)
purch_analy_gendemo_table.head()

```

```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Age Demographics - Analysis
purchase_data["AgeDemo"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

#purchase_data.dtypes
#AgeDemo is a category type instead of integer
purchage_data_df = purchase_data.groupby("AgeDemo")
purchage_data_df.min()

totage_counts = purchage_data_df["AgeDemo"].count()
#totage_counts
totage_per = (totage_counts/Tot_Players)*100

age_demo_df = pd.DataFrame({"Total Count": totage_counts,
    "Percentage of Players": totage_per.map("{:.2f}".format)})
#Reorganize the age demographics
agedemo_table = age_demo_df[["Percentage of Players","Total Count"]]
```

```python
#Age Demographics
agedemo_table
```

```python
#Purchasing Analysis(Age) - Analysis
age_group = purchase_data.groupby("AgeDemo")
purcnt_grpage = age_group["Item Name"].count()
avgpur_grpage = age_group["Price"].mean()
totpur_grpage = age_group["Price"].sum()

purch_analy_agedemo_table = pd.DataFrame({"Purchase Count": purcnt_grpage,
                            "Average Purchase Price": avgpur_grpage.map("${:.2f}".format),
                            "Total Purchase Value": totpur_grpage.map("${:.2f}".format),
                            "Avg Purchase Total per Person": avgpur_grpage.map("${:.2f}".format)})

```

```python
#Purchasing Analysis(Age)
purch_analy_agedemo_table.head()
```

```python
#Top Spenders - Analysis
SN_group = purchase_data.groupby("SN")
purcnt_topspd = SN_group["Item Name"].count()
avgpur_topspd = SN_group["Price"].mean()
totpur_topspd = SN_group["Price"].sum()

top_spenders = pd.DataFrame({"Purchase Count": purcnt_topspd,
                            "Average Purchase Price": avgpur_topspd.map("${:.2f}".format),
                            "Total Purchase Value": totpur_topspd.map("${:.2f}".format)})



top_spenders_summary = top_spenders.sort_values(
    ["Purchase Count", "Average Purchase Price", "Total Purchase Value"], ascending=False)
```

```python
#Top Spenders
top_spenders_summary.head()
```

```python
#Most Popular Items - Analysis
pop_group = purchase_data.groupby(['Item ID', 'Item Name'])
purcnt_pop = pop_group["Item Name"].count()
avgpur_pop = pop_group["Price"].mean()
totpur_pop = pop_group["Price"].sum()

most_pop = pd.DataFrame({"Purchase Count": purcnt_pop,
                            "Average Purchase Price": avgpur_pop.map("${:.2f}".format),
                            "Total Purchase Value": totpur_pop.map("${:.2f}".format)})

most_popular_summary = most_pop.sort_values(
    ["Purchase Count", "Average Purchase Price", "Total Purchase Value"], ascending=False)
```

```python
#Most Popular Items
most_popular_summary.head()
```

```python
#Most Profitable Items - Analysis
most_profitable = most_popular_summary.sort_values(
    ["Purchase Count", "Total Purchase Value", "Average Purchase Price"], ascending=False)

#Reorganize the Most Profitable Items columns
most_profitable_summary = most_profitable[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
```
